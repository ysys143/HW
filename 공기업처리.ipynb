```python
import os
import pandas as pd
```

```python
def list_excel_files(directory_path):
    """
    지정된 디렉토리에서 PDF 파일 목록을 반환합니다.
    """
    pdf_files = []
    if not os.path.isdir(directory_path):
        print(f"오류: '{directory_path}'는 유효한 디렉토리가 아닙니다.")
        return pdf_files
    for filename in os.listdir(directory_path):
        if filename.lower().endswith(".xlsx") and \
           os.path.isfile(os.path.join(directory_path, filename)):
            pdf_files.append(filename)
    return pdf_files
```

```python
base_directory = "/Users/jaesolshin/Documents/GitHub/HW_analysis/공기업"
publics = list_excel_files(base_directory)
publics[0]
```

```python

def make_row3(df: pd.DataFrame) -> pd.Series:
    """
    각 열에 대해 '열이름_0번째값_1번째값' 형태의 값을 가지는 Series를 생성합니다.
    """
    return pd.Series(df.columns 
                    + '_' + df.loc[0, :] 
                    + '_' + df.loc[1, :]
                    + '_' + df.loc[2, :]
                    , index=df.columns)

def insert_row_at_index(df: pd.DataFrame, insert_idx: int, row_values: pd.Series) -> pd.DataFrame:
    """
    DataFrame의 insert_idx 위치에 새로운 행을 추가합니다.
    이미 해당 인덱스가 존재하면, 그 인덱스 이상을 모두 한 칸씩 뒤로 밀어 비워둡니다.
    row_values는 반드시 df.columns와 동일한 인덱스를 가져야 합니다.
    """
    if insert_idx in df.index:
        new_index = []
        for idx in df.index:
            if isinstance(idx, int) and idx >= insert_idx:
                new_index.append(idx + 1)
            else:
                new_index.append(idx)
        df.index = new_index
    df.loc[insert_idx, :] = row_values.values
    df = df.sort_index()
    return df


dir = "/Users/jaesolshin/Documents/GitHub/HW_analysis/공기업/"
df = pd.read_excel(dir + publics[0])

row3 = make_row3(df)
df = insert_row_at_index(df, 3, row3)
df
```

```python
# DataFrame을 엑셀 파일로 저장
def save_df_to_excel(df: pd.DataFrame, file_path: str):
    output_dir = "/Users/jaesolshin/Documents/GitHub/HW_analysis/공기업_processed/"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    df.to_excel(output_dir + file_path, index=False)

save_df_to_excel(df, publics[0])
```

```python
for file in publics:
    row3 = make_row3(df)
    df = insert_row_at_index(df, 3, row3)
    save_df_to_excel(df, file)
```
