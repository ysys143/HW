```python
# PDF 파일 목록 가져오는 함수
def list_excel_files(directory_path):
    """
    지정된 디렉토리에서 PDF 파일 목록을 반환합니다.
    """
    pdf_files = []
    if not os.path.isdir(directory_path):
        print(f"오류: '{directory_path}'는 유효한 디렉토리가 아닙니다.")
        return pdf_files
    for filename in os.listdir(directory_path):
        if filename.lower().endswith(".xlsx") and \
           os.path.isfile(os.path.join(directory_path, filename)):
            pdf_files.append(filename)
    return pdf_files
```

```python
import pandas as pd

def make_row2(df: pd.DataFrame) -> pd.Series:
    """
    각 열에 대해 '열이름_0번째값_1번째값' 형태의 값을 가지는 Series를 생성합니다.
    """
    return pd.Series(df.columns + '_' + df.loc[0, :] + '_' + df.loc[1, :], index=df.columns)

def insert_row_at_index(df: pd.DataFrame, insert_idx: int, row_values: pd.Series) -> pd.DataFrame:
    """
    DataFrame의 insert_idx 위치에 새로운 행을 추가합니다.
    이미 해당 인덱스가 존재하면, 그 인덱스 이상을 모두 한 칸씩 뒤로 밀어 비워둡니다.
    row_values는 반드시 df.columns와 동일한 인덱스를 가져야 합니다.
    """
    if insert_idx in df.index:
        new_index = []
        for idx in df.index:
            if isinstance(idx, int) and idx >= insert_idx:
                new_index.append(idx + 1)
            else:
                new_index.append(idx)
        df.index = new_index
    df.loc[insert_idx, :] = row_values.values
    df = df.sort_index()
    return df

# 사용 예시
row2 = make_row2(df)
df = insert_row_at_index(df, 2, row2)
df
```
